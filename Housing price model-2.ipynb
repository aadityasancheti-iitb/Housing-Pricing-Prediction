```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
df=pd.read_csv('housing.csv.csv')
print(df)
```

# Statistical measures of House-Prices

```python
df.head()
print(f"Mean of prices: {df['price'].mean()}")
print(f"Std of prices: {df['price'].std()}")


# Calculate and print quartiles
quartiles = np.percentile(df['price'], [25, 50, 75])
print(f"25th Percentile (Q1): {quartiles[0]}")
print(f"50th Percentile (Median or Q2): {quartiles[1]}")
print(f"75th Percentile (Q3): {quartiles[2]}")

# Calculate and print skewness and kurtosis
skewness = skew(df['price'])
kurt = kurtosis(df['price'])
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")

print(f"Inter-Quartile range: {quartiles[2]-quartiles[0]}")

bowley_skewness = (quartiles[0] + quartiles[2] - 2 * quartiles[1]) / (quartiles[2] - quartiles[0])
print(f"Bowley coefficient of skewness: {bowley_skewness}")
```


Kurtosis Close to 3: A kurtosis value close to 3 suggests that the distribution of the target variable is similar to a normal distribution in terms of tailedness.

```python
def splittest(data,ratio):
    shuffled=np.random.permutation(len(data))
    size=int(len(data)*ratio)
    test_ind=shuffled[:size]
    train_ind=shuffled[size:]
    return data.iloc[train_ind],data.iloc[test_ind]
```

```python
trainset,testset=splittest(df,0.4)
print("Train size: ",len(trainset))
print("Test size: ",len(testset))
```

# Lasso & Ridge Regression


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error
from math import sqrt

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.2, random_state=42)

# Initialize Lasso and Ridge regression models
lasso_model = Lasso(alpha=10)  # You can adjust the alpha parameter for Lasso
ridge_model = Ridge(alpha=2.2)  # You can adjust the alpha parameter for Ridge

# Fit the models on the training data
lasso_model.fit(X_train, y_train)
ridge_model.fit(X_train, y_train)

# Make predictions on the test data
lasso_predictions = lasso_model.predict(X_test)
ridge_predictions = ridge_model.predict(X_test)

# Calculate RMSE (Root Mean Squared Error) for Lasso and Ridge
lasso_rmse = sqrt(mean_squared_error(y_test, lasso_predictions))
ridge_rmse = sqrt(mean_squared_error(y_test, ridge_predictions))

print(f"Lasso RMSE: {lasso_rmse}")
print(f"Ridge RMSE: {ridge_rmse}")
```
